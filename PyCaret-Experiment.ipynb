```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
# apply some styling
plt.style.use("ggplot")
rcParams['figure.figsize'] = (12, 6)
```

```python
weather_df = pd.read_csv('cleaned_weather.csv')
# read the solar plant generation data
solar_df = pd.read_csv('plant_output.csv')
# drop unused columns
solar_df = solar_df.drop(['Plant', 'Real-time power(kW)', 'Daily feed-in energy(kWh)', 'Energy purchased today(kWh)'], axis=1)

# Both weather and solar df are date aligned
data_df = weather_df
data_df['output'] = solar_df['Daily yield(kWh)']
# Remove entries with known 0 output
data_df = data_df[~(data_df['output'] == 0)]
data_df.drop(data_df.columns[0], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numeric_cols = data_df.select_dtypes(include=[np.number]).columns
data_df = data_df[numeric_cols]
data_df = pd.DataFrame(scaler.fit_transform(data_df), columns=data_df.columns)
data_df.head(5)
```

```python

import mlflow 
mlflow.set_tracking_uri(uri="http://127.0.0.1:5000")

# import RegressionExperiment and init the class
from pycaret.regression import RegressionExperiment
```

```python
exp = RegressionExperiment()
exp.setup(data_df, target = 'output', log_experiment = True, experiment_name='BasicRegression', use_gpu=True, session_id = 123)
```

```python
# compare regression models
best = exp.compare_models()
```

```python
# evaluate trained model
exp.evaluate_model(best)

# predict on hold-out/test set
pred_holdout = exp.predict_model(best)
```

## Feature Engineering
As the max R2 score is 0.489, let's apply `PolynomialFeatures` 

```python
exp_poly = RegressionExperiment()
exp_poly.setup(data_df, target = 'output', log_experiment = True, experiment_name='PolynomialRegression', session_id = 123, polynomial_features=True, polynomial_degree=2, use_gpu=True)
```

```python
# compare regression models
best_poly = exp_poly.compare_models()
```

```python
exp_poly.evaluate_model(best_poly)

# predict on hold-out/test set
pred_holdout_poly = exp_poly.predict_model(best_poly)
```

## Observations
1. Applying PyCaret to the solar output dataset shows `Huber Regressor` exhibits best R2 score of 0.489 among the regression algorithms
2. Using polynomial features actually degraded the accruacy with best R2 score of 0.387

## Feature Selection
Let's apply feature selection to both origianl dataset and polynomial features to see if the accuracy improves.

```python
exp_sel = RegressionExperiment()
exp_sel.setup(data_df, target = 'output', session_id = 123, feature_selection=True, log_experiment=True, experiment_name='FeatureSelection', use_gpu=True)

# compare regression models
best_sel = exp_sel.compare_models()

# evaluate trained model
exp_sel.evaluate_model(best_sel)

# predict on hold-out/test set
pred_holdout_sel = exp_sel.predict_model(best_sel)
```

```python
exp_sel_poly = RegressionExperiment()
exp_sel_poly.setup(data_df, target = 'output', session_id = 123, feature_selection=True, log_experiment=True, experiment_name='PolynomialFeatureSelection', use_gpu=True)

# compare regression models
best_sel_poly = exp_sel_poly.compare_models()

# evaluate trained model
exp_sel_poly.evaluate_model(best_sel_poly)

# predict on hold-out/test set
pred_holdout_sel = exp_sel_poly.predict_model(best_sel_poly)
```

### Observations
- Fature selection does not help improve the accuracy
